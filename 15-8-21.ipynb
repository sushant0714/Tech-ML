```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
d=pd.read_csv("housing.csv")
```

```python
for i in d.columns[:-2]:
    plt.scatter(d[i],d["Price"])
    plt.xlabel(i)
    plt.ylabel("Price")
    plt.title("{} vs {}".format(i,"price"))
    plt.show()
    
```

```python
for i in d.columns[:-2]:
    plt.hist(d[i])
    plt.xlabel(i)
    plt.ylabel("COUNT")
    plt.title("{} vs {}".format(i,"COUNT"))
    plt.show()
```

```python
d.columns
```

```python
x=d[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms','Area Population']]
y=d["Price"]
```

```python
x
```

```python
y
```

```python
x.corrwith(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y, test_size=0.2,random_state=10)
```

```python
ytrain.shape
```

```python
model=LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.score(xtest,ytest)*100
```

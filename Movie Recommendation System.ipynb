```python
import pandas as pd
import numpy as np
```

```python
d1=pd.read_csv('u.data',sep="\t",names="User_id Item_id Rating Timestamp".split())
```

```python
d1
```

```python
col_names=["Item_id","Title","Time","n/a",'Link']+[str(i) for i in range(19)]
d2=pd.read_csv("u.item",sep="|",names=col_names)
```

```python
d2
```

```python
data=pd.merge(d2[["Item_id","Title"]],d1, on="Item_id")
```

```python
data
```

```python
data_p=data.pivot_table(index="User_id",columns="Title",values="Rating")
```

```python
data_p
```

```python

Data=data.groupby("Title")
```

```python
Rating=pd.DataFrame(Data["Rating"].mean())
```

```python
Rating["Count"]=data["Title"].value_counts()
```

```python
Rating
```

```python
Starwars=data_p["Star Wars (1977)"]
Starwars
```

```python
Star_Corr=data_p.corrwith(Starwars)
Star_Corr
```

```python
Star=pd.DataFrame(Star_Corr,columns=["Correlation"])
```

```python
Star
```

```python
Star.dropna(inplace=True)
```

```python
Final=Star.join(Rating["Count"])
```

```python
Final
```

```python
Final[Final['Count']>200].sort_values(by='Correlation',ascending=False).head(30)
```

```python
Lion_ratings = data_p["Lion King, The (1994)"]
similarity_lion = data_p.corrwith(Lion_ratings)
corr_df_lion = pd.DataFrame(similarity_lion, columns = ['Correlation'])
corr_df_lion.dropna(inplace=True)
corr_df_lion = corr_df_lion.join(Rating['Count'])
corr_df_lion[corr_df_lion['Count']>200].sort_values(by='Correlation',ascending=False).head(3)
```

```python
Alddin_rating=data_p["Aladdin (1992)"]
Corr=data_p.corrwith(Alddin_rating)
Cor_lion=pd.DataFrame(Corr,columns=["Correlation"])
Cor_lion.dropna(inplace=True)
Cor_lion=Cor_lion.join(Rating["Count"])
Cor_lion[Cor_lion["Count"]>200].sort_values(by="Correlation",ascending=False).head(10)
```

```python
from sklearn.model_selection import train_test_split
X=df.drop(columns=["crew"])
```

```python
X
```

```python
Y=df["crew"]
```

```python
Y
```

```python
X.corrwith(Y)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.25,random_state=100)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(xtrain,ytrain)
```

```python
lr.coef_
```

```python
lr.score(xtest,ytest)*100
```

```python
ypred=lr.predict(xtest)
```

```python
ypred
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
print(np.sqrt(mean_squared_error(ytest,ypred)))
print(mean_absolute_error(ytest,ypred))
lr_score=print(r2_score(ytest,ypred)*100)
```

The algorithm we have used in this dataset is "Linear Regression" as the crew members is not categorical data. We just have to predict the number of crew members. Linear Regression is best algo to predict models that shows linear relationship.
